# Basic Python Group Project
Duration: 1 hour and 30 minutes

## Instructions

Create a Python program that uses:

- Variables
- Data Types
- Input and Output
- Arithmetic Operators
- Comparison Operators
- Conditional Statements
- Loops

Each group must present their program and explain:

1. What inputs were used
2. What operators were applied
3. What condition was implemented
4. What loop was used

---

# Group 1: Student Grade Evaluator

## Requirements

1. Ask for student name
2. Ask for three grades
3. Compute the average
4. Display remark based on average:
   - 90 and above → Excellent
   - 75 to 89 → Passed
   - Below 75 → Failed
5. Ask if user wants to enter another student
6. Use a loop to repeat the process

# Group 2: Simple ATM Simulator

## Requirements

1. Ask for user name
2. Set initial balance (example: 1000)
3. Display menu:
   1 - Deposit
   2 - Withdraw
   3 - Check Balance
   4 - Exit
4. Use a loop to keep showing the menu
5. Prevent withdrawal if balance is insufficient

# Group 3: Number Guessing Game

## Requirements

1. Set a secret number
2. Ask user to guess
3. If guess is too high → print "Too high"
4. If guess is too low → print "Too low"
5. If correct → display number of attempts
6. Use a loop until the correct guess

# Group 4: Mini Enrollment Fee Calculator

## Requirements

1. Ask for student name
2. Ask number of units
3. Set price per unit (example: 1500)
4. Compute total tuition
5. Apply discount:
   - 21 units and above → 10% discount
   - Below 21 units → No discount
6. Display final amount
