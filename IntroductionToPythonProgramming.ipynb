## Variables in Python

A variable is a container used to store data.

We use variables to:
- Store information
- Reuse values
- Make programs dynamic

---

### Rules for Naming Variables:

1. Must start with a letter or underscore (_)
2. Cannot contain spaces
3. Only letters, numbers, and underscores are allowed
4. Case sensitive (Name and name are different)
5. Cannot use reserved words (like if, for, print)

## Decision Making: if-elif-else

The if-elif-else statement allows a program to make decisions.

The program checks conditions from top to bottom.
Only the first True condition runs.

```python
age = 20

if age >= 18:
    print("You are an adult.")
else:
    print("You are a minor.")
```

## The range() Function

`range()` generates a sequence of numbers.

It is commonly used with a `for` loop.

Basic forms:

range(stop)
range(start, stop)
range(start, stop, step)

Important:
- It starts at the first number.
- It stops BEFORE the last number.

## while Loop

A `while` loop repeats code as long as a condition is True.

Be careful: if the condition never becomes False, the loop will run forever.
